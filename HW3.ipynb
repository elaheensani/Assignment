Write a function `analyze_numbers` that takes two numbers as input. The function should:
- Print both numbers
- Calculate their sum
- Print whether the sum is even or odd

```python
def analyze_numbers(number1, number2): 
    print(number1, number2)
    sum = number1 + number2 
    if sum % 2 == 0:
        print(f"The sum is even")
    else:
        print(f"The sum is odd")  
    return sum
first_number = int(input("enter the first number"))
second_number = int(input("enter the second number"))
analyze_numbers(first_number, second_number)
```

## Cell 4: Function with Default Argument

Read the function below carefully.

üå± Your Answer:The function gets the original price from the user, then by writing two lines of codes, it calculates the discount amount and the price after discount.Then prints both of them. After that, we do the same process for tax and final price and print them. At the end, the def rounds the final price to 2 decimal places and returns it.

```python
def calculate_discounted_price(price, discount=10, tax=5):
    print(f"Original price: ${price}")
    
    discount_amount = price * (discount / 100)
    price_after_discount = price - discount_amount
    print(f"Discount ({discount}%): -${discount_amount}")
    print(f"Price after discount: ${price_after_discount}")
    
    tax_amount = price_after_discount * (tax / 100)
    final_price = price_after_discount + tax_amount
    print(f"Tax ({tax}%): +${tax_amount}")
    print(f"Final price: ${final_price}")
    
    return round(final_price, 2)

# Example usage:
calculate_discounted_price(100)
```

## Cell 7: Exception Handling

Carefully read the following code.

üå± Your Answer: The program asks the user to enter two numbers and then tries to divide them.we use except to handle the error. If the user enters something that isn't a number, it'll handle the "type error". If user enters number 0 for the denominator, it'll handle the "zerodivision error".By writing continue we allow the program to run its course and handle the errors.If the division works, it prints the result rounded to three decimal places.At the end, it always thanks the user, no matter what happens.

```python
while True:
    try:
        numerator = float(input("Enter the numerator: "))
        denominator = float(input("Enter the denominator: "))

        result = numerator / denominator
    except ValueError:
        print("‚ö†Ô∏è Please enter valid numeric values.")
        continue
    except ZeroDivisionError:
        print("‚ö†Ô∏è Denominator cannot be zero. Try again.")
        continue
    else:
        print(f"‚úÖ Result: {round(result, 3)}")
        break
    finally:
        print("üìò Thank you for using the divider.\n")
```
